Use WordNet to identify semantic relationships such as synonymy, antonymy, and hypernymy
in text data.

Importing and Installing Libraries

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
```

**Synonyms**

```python
word_to_analyze = 'good'
synonyms = []
for syn in wordnet.synsets(word_to_analyze):
    for l in syn.lemmas():
        synonyms.append(l.name())

print(f"Synonyms for '{word_to_analyze}':")
for s in sorted(list(set(synonyms))):
    print(f"- {s}")
```

**Antonyms**

```python
word_to_analyze = 'good'
antonyms = []
for syn in wordnet.synsets(word_to_analyze):
    for l in syn.lemmas():
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(f"Antonyms for '{word_to_analyze}':")
for a in sorted(list(set(antonyms))):
    print(f"- {a}")
```

**Hypernyms**

```python
word_to_analyze = 'dog'


dog_synset = wordnet.synsets(word_to_analyze)[0]

print(f"All unique hypernyms (iterative) for '{word_to_analyze}' ({dog_synset.name()}):\n")

hypernyms_list = []
queue = [dog_synset]
seen = set()

while queue:
    current_synset = queue.pop(0)

    if current_synset not in seen:
        seen.add(current_synset)
        for hypernym in current_synset.hypernyms():
            hypernyms_list.append(hypernym.name().split('.')[0])
            queue.append(hypernym)

for hypernym_name in sorted(list(set(hypernyms_list))):
    print(f"- {hypernym_name}")
```
