```python
import string

def preprocess_text(text):

    text = text.lower()


    text_no_punct = ''.join([char for char in text if char not in string.punctuation])


    tokens = nltk.word_tokenize(text_no_punct)


    pos_tags = nltk.pos_tag(tokens)

    return pos_tags

print("The preprocess_text function has been defined.")
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

sample_text = "NLTK is a powerful platform for building Python programs to work with human language data. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and semantic reasoning. NLTK is free, open source, and cross-platform." # Define a sample text


pos_tagged_tokens = preprocess_text(sample_text)


words = [word for word, tag in pos_tagged_tokens]
pos_tags = [tag for word, tag in pos_tagged_tokens]


word_freq_dist = nltk.FreqDist(words)


pos_tag_freq_dist = nltk.FreqDist(pos_tags)


print("\nTop 10 most common words:")
for word, frequency in word_freq_dist.most_common(10):
    print(f"{word}: {frequency}")


print("\nTop 10 most common POS tags:")
for tag, frequency in pos_tag_freq_dist.most_common(10):
    print(f"{tag}: {frequency}")
```

```python
import matplotlib.pyplot as plt


words, word_frequencies = zip(*word_freq_dist.most_common(10))

plt.figure(figsize=(10, 6))
plt.bar(words, word_frequencies, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words')
plt.xticks(rotation=45, ha='right')


tags, tag_frequencies = zip(*pos_tag_freq_dist.most_common(10))

plt.figure(figsize=(10, 6))
plt.bar(tags, tag_frequencies, color='lightcoral')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common POS Tags')

plt.tight_layout()
plt.show()
```
